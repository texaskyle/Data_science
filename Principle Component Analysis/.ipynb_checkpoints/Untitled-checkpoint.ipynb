```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("glass.data")
data
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data = data.drop(columns =['index', 'Class'], axis=1)
```

```python
data
```

```python
data.shape
```

```python
# data is in a diffrent scale, therefore it needs to be standardized
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
scaled_data
```

```python
# i convert the scalled data into a dataframe

df = pd.DataFrame(data = scaled_data, columns = data.columns)
```

```python
df
```

```python
df.describe()
```

```python
from sklearn.decomposition import PCA
pca = PCA()
principleComponents = pca.fit_transform(df)
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Number of Components")
plt.ylabel("Variance (%)")
plt.grid()
plt.title("Explained Variance")
```

from the graph 4 principle component describe almost 90% of the variance in the data and 5 principle components describe almost 95% of the data.

so instead of giving all the columns as inputs, we will only feed this 4 principle componets in the machine learning algorithm
and we will obtain similar results.

```python
pca = PCA(n_components=4)
new_data = pca.fit_transform(df)
# this will be the new data feed into the algorithm
new_data
```

```python
# creating a dataframe of the new data
columns = ['principle_component 1', 'principle_component 2', 'principle_component 3', 'principle_component 4']
principle_DF = pd.DataFrame(data = new_data, columns = columns)
principle_DF
```
