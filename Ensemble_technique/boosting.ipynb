```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.preprocessing import StandardScaler 
from sklearn import datasets
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
import pickle
```

```python
data = pd.read_csv("pima-indians-diabetes.csv")
data
```

```python
data.describe()
```

```python
data.columns
```

in my data i had missing values in crutial columns, the missing values were replaced with zero, 

i want to replace the zeros with nan

```python
cols = ['Plasma glucose concentration',
       'Diastolic blood pressure (mm Hg)', 'Triceps skinfold thickness (mm)',
       '2-Hour serum insulin (mu U/ml)',
       'Body mass index (weight in kg/(height in m)^2)',
       'Diabetes pedigree function', 'Age']
```

```python
data[cols]
```

```python
for col in cols:
    data[cols] = data[cols].replace(0, np.nan)
```

```python
data
```

```python
# checking for missing vakues in my dataset
data.isnull().sum()
```

```python
data['Triceps skinfold thickness (mm)'].mean()
```

```python
# imputing the missing values
data['Plasma glucose concentration'] = data['Plasma glucose concentration'].fillna(data['Plasma glucose concentration'].mode()[0])
data['Diastolic blood pressure (mm Hg)'] = data['Diastolic blood pressure (mm Hg)'].fillna(data['Diastolic blood pressure (mm Hg)'].mode()[0])
data['Triceps skinfold thickness (mm)'] = data['Triceps skinfold thickness (mm)'].fillna(data['Triceps skinfold thickness (mm)'].mean())
data['2-Hour serum insulin (mu U/ml)'] = data['2-Hour serum insulin (mu U/ml)'].fillna(data['2-Hour serum insulin (mu U/ml)'].mean())
data['Body mass index (weight in kg/(height in m)^2)'] = data['Body mass index (weight in kg/(height in m)^2)'].fillna(data['Body mass index (weight in kg/(height in m)^2)'].mean())
```

```python
# checking the values after imputation
data.isnull().sum()
```

```python
# separating the feature and the label column
X = data.drop('Is Diabetic', axis=1)
y = data['Is Diabetic']
```

```python
X.head()
```

```python
# the datapoints differ in magnitude, therefore they will need to be scalled
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# splitting dataset into training and testing dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.30, random_state = 42)
```

```python
# fitting the model the training data
xgboost_model = XGBClassifier(objective='binary:logistic')
xgboost_model
```

```python
xgboost_model.fit(x_train, y_train)
```

```python
# checking the training accuracy of the model
xgboost_model.score(x_train, y_train)
```

```python
# checking the  test accuracy of the model
xgboost_model.score(x_test, y_test)
```

```python
y_pred = xgboost_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

## hyperparameter tunning to increase the accuracy

```python
param_grid={
   
    ' learning_rate':[1,0.5,0.1,0.01,0.001],
    'max_depth': [3,5,10,20],
    'n_estimators':[10,50,100,200]
    
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(XGBClassifier(objective='binary:logistic'),param_grid, verbose=3)
```

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
new_model = XGBClassifier(learning_rate=1, max_depth=5, n_estimators = 50)
```

```python
new_model.fit(x_train, y_train)
```

```python
new_model.score(x_test, y_test)
```

it is clear that the accuracy of the model increased after tunning the hyperparameters
